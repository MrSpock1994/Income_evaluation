```python
income["native-country"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
columns = ["workclass","education", "marital-status","occupation", "occupation", "relationship", "race", "sex","native-country", "income"]
for col in columns:
    income[col] = labelencoder.fit_transform(income[col])
```

```python
income
```

```python
income.drop(["fnlwgt", "capital-gain", "capital-loss", "hours-per-week" ], axis = 1, inplace = True)
```

```python
income.info()
```

```python
x_features = income.loc[:, "age":"native-country"]
```

```python
y_target = income["income"]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_features, y_target, test_size=0.2)
```

```python
model = LogisticRegression(max_iter=100000)
model.fit(x_train, y_train)
```

```python
#Lets apply cross validation
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, x_features, y_target, cv=5)
```

```python
scores
average_score = scores.mean()
average_score
```

```python
predictions = model.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```
